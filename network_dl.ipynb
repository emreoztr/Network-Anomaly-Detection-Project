```python
list_drop = ['id','attack_cat', 'label']
```

```python
df.drop(list_drop,axis=1,inplace=True)
```

```python
numerical_data = df.select_dtypes(include=['number'])
original_feature_count = numerical_data.shape[1]
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.1)
reduced_data = selector.fit_transform(numerical_data)

selected_features = numerical_data.columns[selector.get_support()]
print("Chosen Features:", selected_features)
selected_feature_count = selector.get_support().sum()
```

```python
dropped_feature_count = original_feature_count - selected_feature_count

print(f"Original Feature Count: {original_feature_count}")
print(f"Chosen Feature Count: {selected_feature_count}")
print(f"Dropped Feature Count: {dropped_feature_count}")
```

```python
categorical_data = df.select_dtypes(exclude=['number'])
reduced_data = pd.DataFrame(reduced_data, columns=selected_features)
```

```python

scaled_data_df = reduced_data.apply(np.log1p)
```

```python

final_data = pd.concat([scaled_data_df, categorical_data.reset_index(drop=True)], axis=1)
```

```python
final_data.head()
```

```python
df = final_data
df.describe(include='all')
```

```python
for col in categorical_data:
    print(f"{col}: {df[col].nunique()} different classes")
    print(df[col].value_counts().head(10))

top_6_classes = df['proto'].value_counts().head(6).index

df['proto'] = df['proto'].apply(lambda x: x if x in top_6_classes else 'other')

print(df['proto'].value_counts())
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)

data_encoded = pd.get_dummies(df, columns=categorical_columns)
df = data_encoded
print(data_encoded.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

df_original = pd.read_csv('./drive/MyDrive/unsw nb15/UNSW_NB15_training-set.csv')
X = df
y = df_original['attack_cat']
classes = np.unique(y)
print(classes)
encoder = LabelEncoder()
y = encoder.fit_transform(y)

y_onehot = to_categorical(y, num_classes=10)

X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import Precision, Recall
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

precision_metric = tf.keras.metrics.Precision()
recall_metric = tf.keras.metrics.Recall()

def f1_score(y_true, y_pred):
    precision = precision_metric(y_true, y_pred)
    recall = recall_metric(y_true, y_pred)
    f1 = 2 * ((precision * recall) / (precision + recall + tf.keras.backend.epsilon()))
    return f1

model = Sequential([
    Dense(32, activation='relu', input_dim=X_train.shape[1]),
    Dense(24, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', Precision(name='precision'), Recall(name='recall'), f1_score])
model.fit(X_train, y_train, epochs=2500, batch_size=256, validation_split=0.2, callbacks=[early_stopping])
```

```python
test_loss, test_accuracy, test_precision, test_recall, test_f1 = model.evaluate(X_test, y_test)
print(f"Loss: {test_loss}, Accuracy: {test_accuracy}, Precision: {test_precision}, Recall: {test_recall}")
```
